### (4) We show how $\hat{O}$, the estimated number of discriminable olfactory stimuli, varies using the procedure in <i>trillion</i> as we change the values of incidental parameters.  We repeat each step using the calculation by subjects (as in <i>trillion</i> Fig. 3a) and by mixtures (as in <i>trillion</i> Fig. 3b).

#### Fig. 3: If we vary the number of subjects $S$, or the number of tests $T$:

```python
# By subjects; data in 'subjects_a':
ax_a = plt.subplot(1,2,1)
subjects_a = [trillion.num_odors_vs_replicates(r,fig='a')]

# By tests; data in 'tests_b':
ax_b = plt.subplot(1,2,2)
tests_b = [trillion.num_odors_vs_replicates(r,fig='b')]

# Formatting.  
ax_a.text(0.90, 0.95, 'A', transform=ax_a.transAxes,
      fontsize=16, fontweight='bold', va='top')
ax_b.text(0.90, 0.95, 'B', transform=ax_b.transAxes,
      fontsize=16, fontweight='bold', va='top')
plt.tight_layout()
plt.subplots_adjust(wspace=0.4)
```

#### Fig. 4: If we vary the significance threshold $\alpha$:

```python
# By subjects; data in 'alpha_a':
ax_a = plt.subplot(1,2,1)
alpha_a = [trillion.num_odors_vs_alpha(r,fig='a'),
           trillion.num_odors_vs_alpha(r,fig='a',multiple_correction='fdr')]

# By mixtures; data in 'alpha_b':
ax_b = plt.subplot(1,2,2)
alpha_b = [trillion.num_odors_vs_alpha(r,fig='b'),
           trillion.num_odors_vs_alpha(r,fig='b',multiple_correction='fdr')]

# Formatting.  
ax_a.text(0.90, 0.95, 'A', transform=ax_a.transAxes,
      fontsize=16, fontweight='bold', va='top')
ax_b.text(0.90, 0.95, 'B', transform=ax_b.transAxes,
      fontsize=16, fontweight='bold', va='top')
plt.tight_layout()
plt.subplots_adjust(wspace=0.4)

# Blue is without multiple comparisons correction (as in the authors' paper).  
# Red is with multiple comparisons correction.  
```

#### Fig. 5: If we vary the size of the molecular library $C$:

```python
# By subjects; data in 'C_a':
ax_a = plt.subplot(1,3,1)
C_a = [trillion.num_odors_vs_C(r,fig='a')]

# By mixtures; data in 'C_b':  
ax_b = plt.subplot(1,3,2)
C_b = [trillion.num_odors_vs_C(r,fig='b')]

ax_c = plt.subplot(1,3,3)
c_list = 2**np.arange(5,20)
o_list = [trillion.disc(30,_,30) for _ in c_list]
plt.scatter(c_list,o_list)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Number of components (C) \n in the molecular library")
plt.ylabel(r'Estimated number of discriminable stimuli $\hat{O}$')

# Formatting.  
ax_a.text(0.05, 0.95, 'A', transform=ax_a.transAxes,
      fontsize=16, fontweight='bold', va='top')
ax_b.text(0.05, 0.95, 'B', transform=ax_b.transAxes,
      fontsize=16, fontweight='bold', va='top')
ax_c.text(0.05, 0.95, 'C', transform=ax_c.transAxes,
      fontsize=16, fontweight='bold', va='top')
fig = plt.gcf()
fig.set_size_inches(12,5)
plt.tight_layout()
#plt.subplots_adjust(wspace=0.4)
```

#### Fig. 6: If we vary $S$ (or $T$) and $C$ simultaneously:

```python
# By subjects; data in 's_C_a':
ax_a = plt.subplot(1,2,1)
s_C_a = trillion.num_odors_vs_replicates_and_C(r,fig='a')

# By mixtures; data in 't_C_b':  
ax_b = plt.subplot(1,2,2)
t_C_b = trillion.num_odors_vs_replicates_and_C(r,fig='b')

ax_a = plt.subplot(1,2,1)
plt.pcolor(np.log10(s_C_a[2]))
ax_a.set_xticks(np.arange(0,len(s_C_a[1]))+0.5)
ax_a.set_xticklabels(s_C_a[1],rotation=80);
ax_a.set_yticks(np.arange(0,len(s_C_a[0]))+0.5)
ax_a.set_yticklabels(s_C_a[0]);
plt.xlabel("$C$")
plt.ylabel("$S$")
ax_a_bar = plt.colorbar()
ax_a_bar.set_label("Log$_{10}$ of $\hatO$")

ax_b = plt.subplot(1,2,2)
plt.pcolor(np.log10(t_C_b[2]))
ax_b.set_xticks(np.arange(0,len(t_C_b[1]))+0.5)
ax_b.set_xticklabels(t_C_b[1],rotation=80);
ax_a.set_yticks(np.arange(0,len(t_C_b[0]))+0.5)
ax_b.set_yticklabels(t_C_b[0]);
plt.xlabel("$C$")
plt.ylabel("$T$")
ax_b_bar = plt.colorbar()
ax_b_bar.set_label("Log$_{10}$ of $\hatO$")

# Formatting.  
ax_a.text(1.2, 0.95, 'A', transform=ax_a.transAxes,
      fontsize=16, fontweight='bold', va='top')
ax_b.text(1.2, 0.95, 'B', transform=ax_b.transAxes,
      fontsize=16, fontweight='bold', va='top')
#plt.subplots_adjust(wspace=0.2)
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.tight_layout()
```

#### Fig. 7: If we vary $\alpha$ and $C$ simultaneously:
