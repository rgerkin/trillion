```python
def gilbert_varshamov(C,N,d):
    n = C
    w = N
    d *= 2
    r = d - 1
    combos = trillion.get_n_combos
    num = combos(n,w)
    denom = 0
    for i in range(0,r+1):
        m = min(int(np.floor(i/2)),n-w)
        for j in range(0,m+1):
            denom += combos(w,j)*combos(n-w,j)*combos(w-j,i-2*j)
    return num/denom
```

```python
gilbert_varshamov(128,30,15)
```

```python
def gilbert_varshamov2(C,N,d):
    combos = trillion.get_n_combos
    num = combos(C,N)
    denom = 0
    for i in range(0,d*2):
        m = int(np.floor(i/2))
        for j in range(0,m+1):
            denom += combos(N,j)*combos(C-N,j)*combos(N-j,i-2*j)
    return num/denom
```

```python
def sphere_packing(C,N,d):
    combos = trillion.get_n_combos
    num = combos(C,N)
    denom = 0
    for i in range(0,d):
        denom += combos(N,i)*combos(C-N,i)
    return num/denom
```

```python
def lemma_17(C,N,d):
    n = C
    w = N
    assert (d*2) == (w + 1)
    return factorial(w/2)*(n**(w/2))/factorial(w)
```

```python
def best(C,N,d):
    n = C
    w = N
    d *= 2
    combos = trillion.get_n_combos
    t = int(np.ceil((2*w - d + 1)/2))
    if np.mod(d,2):
        low = combos(n,t)/(4*combos(w,t)*combos(w,t))
        high = combos(n,t)/combos(w,t)
    else:
        low = combos(n,t)/(4*combos(w,t)*combos(w,t)*t)
        high = combos(n,t)/combos(w,t)
    return low,high
```

```python
fnds = np.arange(0.35,1.05,0.05)
Ts = np.ceil(20.0*2.0**np.arange(-2,4,0.5))
appendix_data = np.zeros((len(fnds),len(Ts)))
for i,fnd_ in enumerate(fnds):
    for j,T_ in enumerate(Ts):
        val = 0
        for k in range(0,1+np.round(fnd_*T_).astype('int')):
            val += trillion.get_n_combos(T_,k)*((1/3)**k)*((2/3)**(T_-k))
        if val > 1:
            val = 1.0 # In the case of very high T, Stirling's approximation breaks down and a minor correction is required.  
        appendix_data[i,j] = (1-val)/2
```
