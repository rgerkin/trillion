#### Fig. 8: Edit the values in the first section, to regenerate <i>trillion</i>'s Fig. 4c and 4d.   Note that, as in the original </i>trillion</i>, the left most part of each panel (0%) results in the smallest value for $\hat{O}$ allowed by their framework (which, oddly, is >> 1), and the right most part of each panel (100%) results in the largest value for $\hat{O}$ allowed by their framework (which is all possible mixtures, i.e. C-choose-N.  

```python
# Edit the parameters in this section.  
C = 128 # Size of molecular library.  
alpha = 0.05 # Signficance threshold.  
S = 26 # Number of subjects.  
T = 20 # Number of tests (mixture pairs) per class.  
multiple_correction = False
factor = 1 # Set to 2 to use formulas correctly, or to 1 to use formulas as in their original paper.  

# This section generates the results.  
D_c = trillion.fig3x(results,'a',alpha=alpha,multiple_correction=False,n_replicates=T,plot=False)
D_d = trillion.fig3x(results,'b',alpha=alpha/2,multiple_correction=False,n_replicates=S,plot=False)
# For alpha/2 explantion, see note for Fig. 2b (their Fig. 3b).  
x10 = np.arange(0,11)
x20 = np.arange(0,21)
x30 = np.arange(0,31)
disc = trillion.disc

# This section generates the figure.  
for factor in [1,2]:
    for i,(panel,D) in enumerate({'c':D_c,'d':D_d}.items()):
        ax = plt.subplot(2,2,(factor-1)*2+i+1)
        D_f = (100-D)/100 # Express as fraction distinct components.  
        y10 = [disc(10,C,factor*_) for _ in x10]
        y20 = [disc(20,C,factor*_) for _ in x20]
        y30 = [disc(30,C,factor*_) for _ in x30]
        plt.plot(100*(1-x10/10),y10,color='black',linewidth=2)
        plt.plot(100*(1-x20/20),y20,color='magenta',linewidth=2)
        plt.plot(100*(1-x30/30),y30,color='green',linewidth=2)
        plt.plot([D,D],[1,disc(30,C,factor*30*D_f)],'k')
        plt.plot([0,D],[disc(30,C,factor*30*D_f),disc(30,C,factor*30*D_f)],'k')
        plt.plot([0,D],[disc(20,C,factor*20*D_f),disc(20,C,factor*20*D_f)],'k')
        plt.plot([0,D],[disc(10,C,factor*10*D_f),disc(10,C,factor*10*D_f)],'k')
        plt.yscale('log')
        plt.xlabel("% mixture overlap \n allowing discrimination");
        plt.ylabel(r'Estimated number of discriminable stimuli $\hat{O}$');
        ax.text(0.03, 0.95, chr(ord('A')+i+2*(factor-1)), transform=ax.transAxes,
          fontsize=16, fontweight='bold', va='top')

fig = plt.gcf()
fig.set_size_inches(10,8)
plt.tight_layout()
```

#### "If 50 additional anosmic individuals were enrolled in the study..."

```python
# Note this cell will produce a new result each time it is executed, since it is stochastic calculation.  
from numpy.random import binomial
# Create 234 new anomsics, each taking the same tests and getting each correct according to chance (1/3).  
anosmic_results = [trillion.Result(_.test,_.subject_id,binomial(1,1.0/3)) for _ in results]*9
# Add these to the existing subjects.  
T=20
factor = 1
results2 = results + anosmic_results
# Compute new D's.  
for r in [results,results2]:
    S=26.0*len(r)/len(results) # New number of subjecsts.  
    print("Actual 26 subjects:" if r==results else "Actual 26 subjects + 234 anosmics:")
    for fig in ['a','b']:
        for N in [10,20,30]:
            alpha = 0.05/(2 if fig == 'b' else 1)
            n_replicates = T if fig =='a' else S
            D = trillion.fig3x(r,fig,alpha=alpha,multiple_correction=False,n_replicates=n_replicates,plot=False)
            D_f = (100-D)/100 # Express as fraction distinct components.
            #print(fig,alpha,n_replicates,D_f)
            O_ = trillion.disc(N,C,factor*N*D_f)
            O_max = trillion.get_n_combos(C,N)
            print(u"\tFor Fig %s, N=%d, O = %d, O_relative=%.9g" % (fig,N,O_,O_/O_max))
```
